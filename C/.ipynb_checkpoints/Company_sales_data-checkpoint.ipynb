```python
from pandas import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = read_csv('company_sales_data.csv')
df
```

### <font color = 'green'>Exercício 1</font>
Ler o lucro total (**Total Profit**) de todos os meses e mostre-os através de um gráfico de linhas.  
Usar o dado fornecido do lucro total para cada mês. A gráfico de linha gerado deve incluir as seguintes propriedades:  
X label name = Month Number  
Y label name = Total profit

```python
x=df.month_number
y=df.total_profit
plt.plot(x,y)
plt.xticks(x, x, rotation='horizontal')
plt.xlabel('Month Number')
plt.ylabel('Profit in Dollar')
plt.title("Company profit per month")
plt.show()#
```

### <font color = 'green'>Exercício 2</font>  
Ler o lucro total (Total Profit) de todos os meses e apresente-os num gráfico de linhas, com as características indicadas.  
A linha gerada deve incluir as seguintes propriedades de estilo:  
 Estilo de linha pontilhado e cor da linha devem ser vermelhos  
 Mostrar legenda no local inferior direito  
 Label do rótulo X = Número do mês  
 Label do rótulo Y = Número de unidades vendidas  
 Adicione um marcador de círculo  
 Cor do marcador de linha conforme lida  
 A largura da linha deve ser 3  

```python
plt.figure(figsize=(8,5))

x=df.month_number
y=df.total_profit

plt.plot(df.month_number, df.total_profit, 'r--.', markersize=9, marker="o", markerfacecolor='black', linewidth=3, label="Profit data of last year")

plt.title("Company profit per month")
plt.xlabel('Month Number')
plt.ylabel('Profit in Dollar')

plt.legend(["Profit data of last year"], loc='lower right')
plt.show()
```

### <font color = 'green'>Exercício 3</font>  
Mostrar as vendas mensais de todos os produtos num gráfico de
múltiplas linhas, ou seja, uma linha diferente para cada produto.

```python
plt.figure(figsize=(8,5))

#facecream	facewash	toothpaste	bathingsoap	shampoo	moisturizer

x=df.month_number
y=df.total_profit

plt.title("Sales Data")
plt.xlabel('Month Number')
plt.ylabel('Sales Units in number')


p1 = plt.plot(df.facecream, '.-', markersize=12, linewidth=3, label="Face Cream")
p2 = plt.plot(df.facewash, '.-', markersize=12, linewidth=3, label="Face Wash")
p3 = plt.plot(df.toothpaste, '.-', markersize=12, linewidth=3, label="Toothpaste")
p4 = plt.plot(df.bathingsoap, '.-', markersize=12, linewidth=3, label="Bathing Soap")
p5 = plt.plot(df.shampoo, '.-', markersize=12, linewidth=3, label="Shampoo")
p6 = plt.plot(df.moisturizer, '.-', markersize=12, linewidth=3, label="Moisturizer")

plt.legend()
plt.show()
```

### <font color = 'green'>Exercício 4</font>  
Calcular para cada produto o total das vendas do ano e mostre-os usando Pie Chart (gráfico de queijo).  
No Pie chart inscrever a % de vendas de cada produto.


```python
sum_column = df.sum(axis=0)
TotalSalesbyprod = sum_column
#facecream	facewash	toothpaste	bathingsoap	shampoo	moisturizer

TotalSalesFC = (TotalSalesbyprod.facecream/TotalSalesbyprod.total_units)
TotalSalesFW = (TotalSalesbyprod.facewash/TotalSalesbyprod.total_units)
TotalSalesTP = (TotalSalesbyprod.toothpaste/TotalSalesbyprod.total_units)
TotalSalesBS = (TotalSalesbyprod.bathingsoap/TotalSalesbyprod.total_units)
TotalSalesSH = (TotalSalesbyprod.shampoo/TotalSalesbyprod.total_units)
TotalSalesMO = (TotalSalesbyprod.moisturizer/TotalSalesbyprod.total_units)

data = [TotalSalesFC,TotalSalesFW,TotalSalesTP,TotalSalesBS,TotalSalesSH,TotalSalesMO]
label = ["Face Cream","Face Wash","Tooth Paste","Bathing Soap","Shampoo","Moisturizer"]

plt.axis("equal")
plt.pie(data,labels = label,radius=1.5,autopct = "%0.1f%%",shadow=False, startangle=45)


plt.show()
```
