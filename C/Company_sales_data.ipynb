### <font color = 'green'>Exercício 1</font>
Ler o lucro total (**Total Profit**) de todos os meses e mostre-os através de um gráfico de linhas.

X label name = Month Number

```python
x=df.month_number
y=df.total_profit
plt.plot(x,y)
plt.xticks(x, x, rotation='horizontal')
plt.xlabel('Month Number')
plt.ylabel('Profit in Dollar')
plt.title("Company profit per month")
plt.show()#
```
